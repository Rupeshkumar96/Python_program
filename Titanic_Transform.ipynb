```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the dataset
data=sns.load_dataset('titanic')
```

```python
# check the null value is present or not
data.isnull().sum()
```

```python
data
```

```python
# Fill all null value with mean of age using SimpleInputer class.
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean')
data['age']=imp.fit_transform(data[['age']])
```

```python
data['age'].isnull().sum()
```

```python
# Drop unwanted data
data.drop(['deck','sibsp','parch','who','adult_male','alone','class'],axis=1)
```

```python
data
```

```python
# Convert categorial data in numeric help of LableEncoder class.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['alive']=le.fit_transform(data['alive'])
data['sex']=le.fit_transform(data['sex'])
```

```python
data
```

```python
data.drop(['embarked'],axis=1,inplace=True)
```

```python
data
```

```python
num_types=data.select_dtypes(include=np.number)
for num in num_types:
    q1=data[num].quantile(0.25)
    q3=data[num].quantile(0.75)
    iqr=q3-q1
    l1=q1-1.5*iqr
    r1=q3+1.5*iqr
    outlier=data[(data[num]<l1)|(data[num]>r1)].index
```
