```python
# Question8. WAP to find the fibonaci of given number
num=int(input("Enter a number"))
a=-1
b=1
while num>0:
    c=a+b
    print(c,end=' ')
    a=b
    b=c
    num=num-1
```

```python
# WAP to print the factorial of given number
num=int(input("Enter a number"))
t=num
fact=1
while num>0:
    fact=fact*num
    num=num-1;
print(f"The factoria of {t} is: ",fact)
```

```python
# Question 9. WAP to print the n number of series for ex:- n=5 =>2+22+222+2222+22222=24690
n=int(input("Enter a number"))
num=5
s=n
sum=0
while num>0:
    print(n,end='+')
    sum=sum+n
    n=n*10+s
    num=num-1
print("=",sum)
```

```python
# Question 10. WAP to print the prime number between the two number .
start=int(input("Enter a First number:"))
last=int(input("Enter the last number:"))
prime=[]
for i in range(start,last+1):
    n=i//2  # also use sqrt(n)
    for j in range(2,n+1):
        if i%j==0:
            break
    else:
        prime.append(i)
print(f"The Prime number is between {start} to {last} is: ",prime)
```
