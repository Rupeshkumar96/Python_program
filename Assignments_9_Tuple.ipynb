## Problem on Tuples

```python
# Question 1) Find out Sum of tuple elements
# Input(7, 8, 9, 1, 10, 7)
# Output=42

nums=tuple(int(x) for x in input("Ener a number").split(','))
sum=0
for i in nums:
    sum+=i
print(sum)
```

```python
# Question 2) Write a Python program to find repeated items in a tuple.
# Input (2, 4, 5, 6, 2, 3, 4, 4, 7, 4)
# Output=2,4

nums=tuple(map(int,input("Enter a number").split(',')))
lis=[]
nums=list(nums)
for i in nums:
    if nums.count(i)>1 and i not in lis:
        lis.append(i)
print(tuple(lis))
```

```python
# Question 4) Write a Python program to remove an item "p" from a tuple.
# Input ("a", 5, "r", "l", "p", "u", "r", "m", "e")
# Output ('a', 5, 'l', 'u', 'r', 'm', 'e')
nums=tuple(s for s in input("Enter a element").replace('',' ').split())
nums_lis=list(nums)
nums_lis.remove('p')
nums=tuple(nums_lis)
nums
```
