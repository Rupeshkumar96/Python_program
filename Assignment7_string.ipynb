```python
# Question 2  Check if a string has at least one letter and one number
# Input: welcome2ourcountry34
# Output: True

sts=input("Enter a string")
def check_string_digit(sts):
    check_string=False
    check_digit=False
    for i in sts:
        if i.isalpha():
            check_string=True
        if i.isdigit():
            check_digit=True
    return check_string and check_digit
check_string_digit(sts) # Function call always after the declaration.
```

```python
# Question 5 Calculate the sum and average of the digits present in a string
# Input: input_str = "Datapython29@8496"
# Output:
# Sum is: 38
# Average is: 6.333333333333333

string=input("Enter a stirng:")
sum1=0
length=0
for i in string:
    if i.isdigit():
        sum1+=int(i)
        length+=1
print("The Average of the digit which is present in the string:",sum1/length)
```

```python
# Question 6 Count all letters, digits, and special symbols from a given string
# Input: "A@un2st&#5ve"
# Output:
# Chars = 8
# Digits = 2
# Symbol = 3

string=input("Enter a string")
def check_char(string):
    digit=0
    char=0
    special=0
    for i in string:
        if i.isalpha():
            char+=1
        elif i.isdigit():
            digit+=1
        else:
            special+=1
    print(f'In this string the Char is present:{char} and Digit:{digit} and the special char is:{special}')
check_char(string)
```
