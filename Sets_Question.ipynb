## Problem on sets

```python
# 1) Write a Python program to create an intersection of sets.
# Example:
# Input:
# {'green', 'blue'}
# {'blue', 'yellow'}
# Output:
# {'blue'}

ele=set(x for x in input("Enter a Element:").replace(',',' ').split())
ele1=set(x for x in input("Enter a Element:").replace(',',' ').split())
result=ele.intersection(ele1)
print(result)
```

```python
# 2) Write a Python program to create a union of sets.
# Example:
# Input:
# {'blue', 'green'}
# {'blue', 'yellow'}
# Output:
# {'blue', 'yellow', 'green'}

ele=set(s for s in input("Enter a Element:").replace(',',' ').split())
ele1=set(s for s in input("Enter a Element:").replace(',',' ').split())
result=ele.union(ele1)
print(result)
```

```python
# 3) Write a Python program to remove all elements from a given set.
# Example:
# Input:
# {'Green', 'Black', 'Red', 'White'}
# Output:
# set()

ele=set(s for s in input("Enter a Element:").replace(',',' ').split())
print(ele.clear())
```

```python
# 4) Write a Python program to count number of vowels using sets in a given string.
# Example:
# Input:
# str = "helloeveryone"
# Output:
# No. of vowels: 6

ele=input("Enter a Element:")
print(ele)
vowel={'a','e','i','o','u'}
count=0
for i in ele:
    if i in vowel:
        count+=1;
print(f'Total vowel is present in the string is:{count}')
```

```python
#### 5) Find the elements in a given set that are not in another set.
# Example:
# Input:
# {1, 2, 3, 4, 5}
# {4, 5, 6, 7, 8}
# Output:
# Difference of set1 and set2:
# {1, 2, 3}
# Difference of set2 and set1:
# {6, 7, 8}


ele=set(s for s in input("Enter a Element:").split())
ele1=set(s for s in  input("Enter a Element:").split())
result=ele.difference(ele1)
result1=ele1.difference(ele)
print(result)
print(result1)
```

```python
# 6) Write a Python program to add member(s) to a set.
# Example:
# Input:
# set()
# Output after adding single element:
# {'Red'}
# Output after adding multiple elements:
# {'Red', 'Green', 'Blue'}

set_ele=set()
s=input("Enter a Element:").split()
set_ele.update(s)
print(set_ele)

  
```

```python
# 7) Check if two given sets have no elements in common.
# Example:
# Input:
# x = {1, 2, 3, 4}
# y = {4, 5, 6, 7}
# z = {8}
# Output:
# Compare x and y: False
# Compare x and z: True

ele=set(map(int,input("Enter a number").split()))
ele1=set(map(int,input("Enter a number:").split()))
result=ele.issubset(ele1)
print("The number is:",result)
```

```python
# 8) Write a Python program to find the maximum and minimum values in a set.

# Example:
# Input:
# {2, 3, 20, 5, 10, 15}
# Output:
# Maximum value of the said set: 20
# Minimum value of the said set: 2
ele=set(map(int,input("Entera  Element:").split()))
maximum=max(ele)
minimum=min(ele)
print(maximum)
print(minimum)
```
