```python
# 1) Write a Python program to create an intersection of sets.
# Example:
# Input:
# {'green', 'blue'}
# {'blue', 'yellow'}
# Output:
# {'blue'}

ele=set(x for x in input("Enter a Element:").replace(',',' ').split())
ele1=set(x for x in input("Enter a Element:").replace(',',' ').split())
result=ele.intersection(ele1)
print(result)
```

```python
# 2) Write a Python program to create a union of sets.
# Example:
# Input:
# {'blue', 'green'}
# {'blue', 'yellow'}
# Output:
# {'blue', 'yellow', 'green'}

ele=set(s for s in input("Enter a Element:").replace(',',' ').split())
ele1=set(s for s in input("Enter a Element:").replace(',',' ').split())
result=ele.union(ele1)
print(result)
```

```python
# 3) Write a Python program to remove all elements from a given set.
# Example:
# Input:
# {'Green', 'Black', 'Red', 'White'}
# Output:
# set()

ele=set(s for s in input("Enter a Element:").replace(',',' ').split())
print(ele.clear())
```

```python
# 4) Write a Python program to count number of vowels using sets in a given string.
# Example:
# Input:
# str = "helloeveryone"
# Output:
# No. of vowels: 6

ele=input("Enter a Element:")
print(ele)
ele=set(ele)
vowel={'a','e','i','o','u'}
result=vowel.intersection(ele)
print(f'The Length of the Vowel is in given string:{len(result)}')
```
