```python
# Question 1). Python program to interchange first and last elements in a list.
# Input: [12, 35, 9, 56, 24]
# Output: [24, 35, 9, 56, 12]

nums=list(map(int,input("Enter a number").split(' ')))
temp=nums[0]
nums[0]=nums[-1]
nums[-1]=temp
nums
    
```

```python
# Question 2. Find sum and average of List in Python.
# Input: [4, 5, 1, 2, 9, 7, 10, 8]
# Output: sum = 46, average = 5.75
nums=map(int,input("Enter a number").split(" "))
summ=0
count=0
for i in nums:
    summ+=i
    count+=1
average=summ/count
print(f'The sum of number is: {summ} and the average is: {average}')
```

```python
# Question 3). Python program to find second largest number in a list.
# Input: list2 = [70, 11, 20, 4, 100]
# Output: 70

nums=list(map(int,input("Enter a number").split(' ')))
nums.sort(reverse=True)
print(f'The highest number in the list is:{nums[1]}')
```

```python
# Question 4) Python program to count positive and negative numbers in a list.
# Input: [10, -21, 4, -45, 66, -93, 1]
# Output: Positive numbers in the list: 4, Negative numbers in the list: 3
nums=list(map(int,input("Enter a number").split(',')))
positive_count=0
negative_count=0
for i in nums:
    if i<0:
        negative_count+=1
    else:
        positive_count+=1
print(f'The negative number in list is: {negative_count} and the Positive number is: {positive_count}')
```

```python
# 8) How to count unique values inside a list.
# Input: [1, 2, 2, 5, 8, 4, 4, 8]
# Output: No of unique items are: 5

nums=[int(x) for x in input("Enter a list").split(',')]
lis=[]
count=0
for i in nums:
    if i not in lis:
        count+=1
        lis.append(i)
print("Number of unique values are:",count)
```

```python
# Question 9). Append a list to the second list.
# Input: list1 = [1, 2, 3, 0], list2 = ['Red', 'Green', 'Black']
# Output: [1, 2, 3, 0, 'Red', 'Green', 'Black']

list1=[int(x) for x in input("Enter a number").split(',')]
list2=[int(x) for x in input("Enter a number").split(',')]
list1.extend(list2)
list1
```

```python
# Question 10). Write a Python program to replace the last element in a list with another list.
# Sample data: [1, 3, 5, 7, 9, 10], [2, 4, 6, 8]
# Expected Output: [1, 3, 5, 7, 9, 2, 4, 6, 8]

list1=[int(x) for x in input("Enter a number").split(',')]
list2=[int(x) for x in input("Enter a number").split(',')]
list1[-1:]=list2
list1
```
